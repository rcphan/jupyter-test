```python
def image_intersect_area(cntA, cntB, height = 3000, width = 3000):
    try:
        canvasA = np.zeros((height, width), 'uint8')
        canvasB = canvasA.copy()
        canvasAreaValue = []

        canvasToDraw = [canvasA, canvasB]
        cntsToDraw = [cntA, cntB]
        
        print(len(cntsToDraw))

        for cdx, cnt in enumerate(cntsToDraw):
            canvasAreaValue.append(cv2.contourArea(cnt))
            cv2.drawContours(canvasToDraw[cdx], cntsToDraw, cdx, (255,255,255), -1)
#             mc.displayImage(canvasToDraw[cdx])

        canvasIntersect = cv2.bitwise_and(canvasA, canvasB)
        canvasIntersectCnts, _ = cv2.findContours(canvasIntersect, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if len(canvasIntersectCnts) > 0:
            canvasAreaValue.append(cv2.contourArea(canvasIntersectCnts[0]))
            return canvasAreaValue[2]/canvasAreaValue[0]
        else:
            return 0
        
    except Exception as ex:
        print(traceback.format_exc())
```

```python
for ncdx, nct in enumerate(nctL[:10]):
    for mcdx, mct in enumerate(mctL[:10]):
        if image_intersect_area(nct, mct) > 0.7:
            print('inside')
```
